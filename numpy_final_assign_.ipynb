# Q1. Array Creation and Attributes

## Problem Statement
Create array [10,20,30,40,50] and print shape, ndim, size and dtype.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
import numpy as np
arr = np.array([10,20,30,40,50])
print(f"shape of array : {arr.shape}")
print(f"Dimensional of array : {arr.ndim}")
print(f"size of array : {arr.size}")
print(f"Data type of array : {arr.dtype}")
````


# Q10. Concatenation and Stacking

## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])
print(np.vstack((arr1,arr2)))  # vertical stack means ek ke upar ek rakhna
# Top to bottom

print(np.concatenate((arr1,arr2)))  # concatenate me dono array ko jor kar ek
# single array bana deta hai
````
